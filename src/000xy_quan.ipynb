```python
def stretch_ratio(om, w, h, z, om_dft=omdft, w_dft=wdft, h_dft=0.7):
    ratio_dft = 1.0 / (comov_r(om_dft, w_dft, h_dft, z) * Hz(om_dft, w_dft, h_dft, z))
    ratio_ass = 1.0 / (comov_r(om, w, h, z) * Hz(om, w, h, z))
    return ratio_ass / ratio_dft

def vol_ratio(om, w, h, z, om_dft=omdft, w_dft=wdft, h_dft=0.7 ):
    ratio_dft = (comov_r(om_dft, w_dft, h_dft, z)**2.0 / Hz(om_dft, w_dft, h_dft, z))
    ratio_ass = (comov_r(om, w, h, z)**2.0 / Hz(om, w, h, z))
    return ratio_ass / ratio_dft

X = np.linspace(0.01,2.5,101)


fig, ax = figax(8,8)
lws= [4,4,5,6,7]
lss= ['-', '--', ':', '-.', '-.']
lcs= ['k', 'g', 'c', 'r', 'b']
iomw = 0
for omw in cosmos:
    om, w = omw
    Y = [vol_ratio(om,w,0.7,z) for z in X]
    ax.plot(X,Y, lw=lws[iomw], ls=lss[iomw],c=lcs[iomw], label=omwtex(om,w))
    iomw += 1  
ax.legend(loc='best', frameon=False, fontsize=18)
ax.grid()
ax.set_xlabel('$z$',fontsize=18)
ax.set_ylabel('$ V_{\\rm wrong}\\ /\\ V_{\\rm true}$',fontsize=18)
fig.savefig('fig1_b.eps',format='eps')
```
